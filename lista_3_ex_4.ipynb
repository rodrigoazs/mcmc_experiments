```python
import random
import math
import numpy as np
import matplotlib.pyplot as plt

def dist_exp(x, lambda_):
    return lambda_ * math.exp(-lambda_*x)

def dist_exp_mirror(x, lambda_):
    return lambda_ * math.exp(-lambda_*x) if x > 0 else lambda_ * math.exp(-lambda_*(-x))

def dist_normal(x):
    return 1/math.sqrt(2*math.pi) * math.exp(-x**2/2)
```

```python
x = np.arange(-3, 3, 0.1)

v_exp_func = np.vectorize(dist_exp)
y_exp = v_exp_func(x, 1/2)

#v_exp_func = np.vectorize(dist_exp)
#y_exp2 = v_exp_func(x, 1.1)

v_normal_func = np.vectorize(dist_normal)
y_normal = v_normal_func(x)
```

```python
plt.plot(x, y_exp)
#plt.plot(x, y_exp2)
plt.plot(x, y_normal)
plt.show()
```

```python
v_exp_mirror_func = np.vectorize(dist_exp_mirror)
y_exp_mirror = v_exp_mirror_func(x, 1/2)

plt.plot(x, y_exp_mirror)
#plt.plot(x, y_exp2)
plt.plot(x, y_normal)
plt.show()
```

```python
def von_neumann(beta_):
    i = None
    while(i == None or u >= qi):
        i = np.random.exponential(beta_)
        cp = dist_exp(i, 1/beta_)
        u = np.random.uniform(0, cp)
        qi = dist_normal(i)
    return (-1)**(random.randint(0,1))*i
```
