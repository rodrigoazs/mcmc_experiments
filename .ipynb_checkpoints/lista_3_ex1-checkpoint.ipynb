```python
import random
import math
import numpy as np
import matplotlib.pyplot as plt

def dist_beta():
    s = 0
    n = 0
    while(s < 1):
        s += random.random()
        n += 1
    return n

# sigma2 = 0.76
def variance_numeric(n):
    a = []
    for i in range(n):
        a.append(dist_beta())
    a = np.array(a)
    return np.std(a)**2

def variance(n):
    return -(math.exp(1)-3)*math.exp(1)

def sem(n):
    return abs(np.sqrt(variance(n))/np.sqrt(n))

def estimate(n):
    s = 0
    for i in range(n):
        s += dist_beta()
    return s/n

def relative_error(e):
    return abs(e - math.exp(1))/math.exp(1)
```

```python
x = np.arange(1,10**6,10000)
```

```python
en = []
for i in x:
    e = estimate(i)
    en.append(e)
```

```python
plt.plot(x, en)
plt.show()
```

```python
error = []
for i in range(len(x)):
    t = relative_error(en[i])
    error.append(t)
```

```python
plt.plot(x, error)
plt.show()
```

```python
x_log = np.log(x)
y_log = np.log(np.array(error))
plt.plot(x_log, y_log)

z = [sem(i) for i in x]
z_log = np.log(z)
plt.plot(x_log, z_log)

plt.show()
```
