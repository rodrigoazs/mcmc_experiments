```python
import math
import numpy as np
import random
import matplotlib.pyplot as plt
import time

exact = 0
for i in range(1, 1001):
    exact += i * math.log(i)
    
K = 0
for i in range(1, N+1):
    K += i**(1.2)
    
vec = []
for i in range(1, N+1):
    vec.append(i**(1.2)/K)
    
summ = np.sum(vec)

vec = vec/summ

def relative_error(x):
    return abs(x - exact)/exact
```

```python
def experiment(n):
    s = 0
    N = 1000
    for i in range(n):
        x = random.randint(1, N)
        g = x * math.log(x)
        s += g

    return s / n * N
```

```python
def importance_experiment(n):
    s = 0
    N = 1000
    for i in range(n):
        x = np.random.choice(N, p=vec) + 1 #random.randint(1, N)
        g = x * math.log(x)
        h = x**(1.2) / K
        s += g/h

    return s / n
```

```python
x = np.arange(0,10**6+10000,10000)
x[0] = 1
```

```python
error_regular = []
error_importance = []
start_time = time.time()
for i in range(len(x)):
    e = experiment(x[i])
    s = importance_experiment(x[i])
    error_regular.append(relative_error(e))
    error_importance.append(relative_error(s))
elapsed_time = time.time() - start_time
time.strftime("Tempo de execucao: %H:%M:%S", time.gmtime(elapsed_time))
```

```python
plt.plot(np.log(x), np.log(error_regular), label='Sem importance sampling')
plt.plot(np.log(x), np.log(error_importance), label='Com importance sampling')
plt.legend()
plt.show()
```
