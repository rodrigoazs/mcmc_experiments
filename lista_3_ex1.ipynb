```python
import random
import math
import numpy as np
import matplotlib.pyplot as plt

def urnas_vazias(n=10000):
  urna_1 = set()
  for i in range(n):
    j = random.choice(range(n))
    if j not in urna_1:
      urna_1.add(j)
  return len(urna_1)

def variance_numeric(urnas, n):
    a = [1] * urnas
    b = [0] * (n-urnas)
    a.extend(b)
    c = np.array(a)
    return np.std(c)**2

def variance(n):
    return (1-1/n)**n * (1-(1-1/n)**n)

def sem(n):
    return np.sqrt(variance(n))/np.sqrt(n)

def estimate(n):
    u = urnas_vazias(n)
    e = n/(n-u)
    return e

def relative_error(e):
    return abs(e - math.exp(1))/math.exp(1)
```

```python
x = np.arange(1,10**6,10000)
```

```python
en = []
for i in x:
    e = estimate(i)
    en.append(e)
```

```python
plt.plot(x, en)
plt.show()
```

```python
error = []
for i in range(len(x)):
    t = relative_error(en[i])
    error.append(t)
```

```python
x_log = np.log(x)
y_log = np.log(np.array(error))
plt.plot(x_log, y_log)

z = [sem(i) for i in x]
z_log = np.log(z)
plt.plot(x_log, z_log)

plt.show()
```
